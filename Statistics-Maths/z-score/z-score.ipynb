```python
import pandas as pd
```

```python
df = pd.read_csv("heights_with_outliers.csv")
```

```python
mean = df.height.mean()
std = df.height.std()
```

```python
df["z_score"] = (df.height - mean)/std
```

```python
df.head()
```

```python
df_new = df[(df["z_score"] < -3) | (df["z_score"] > 3)]
```

```python
df_new #These are outliers
```

```python
df_clean= df[(df["z_score"] > -3) & (df["z_score"] < 3)]
```
