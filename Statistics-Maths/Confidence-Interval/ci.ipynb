```python
sample_size = 50
sample = df.sample(50, random_state=100)
sample.head()
```

```python
population_mean = df.miles.mean()
population_std = df.miles.std()
population_mean, population_std
```

```python
sample_mean = sample.miles.mean()
sample_std = sample.miles.std()
sample_mean, sample_std
```

```python
from math import *

standard_error = population_std/sqrt(sample_size)
standard_error
```

```python
import scipy.stats as st

# .975 is for 95% confidence level: 95% + 2.5%
z = st.norm.ppf(.975)
z
```

```python
margin_of_error = z * standard_error
```

```python
sample_mean - margin_of_error, sample_mean + margin_of_error
```

```python
print("Confidence interval is: ", round(sample_mean - margin_of_error, 2), "--->", round(sample_mean + margin_of_error, 2))
```
