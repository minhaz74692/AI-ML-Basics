```python
import  pandas as pd
```

```python
df = pd.read_csv("csv-excel-files/heights.csv")
```

```python
df
```

```python
q1, q3 = df.height.quantile([.25, 0.75])
```

```python
IQR = q3-q1
IQR
```

```python
lower_limit = q1-1.5*IQR
upper_limit = q3 + 1.5*IQR
```

```python
print(q1,q3, IQR, lower_limit, upper_limit)
```

```python
df_new = df[(df.height>lower_limit) & (df.height<upper_limit)]
df_new.head()
```
