```python
import pandas as pd
```

```python
df = pd.read_csv("heights.csv")
df.describe()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df.height, color="white", kde=True)
```

```python
mean = df['height'].mean()
std = df['height'].std()
round(mean, 2), round(std,2 )
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    df['height'],
    bins=30,
    kde=True,
    stat='density',
    color='skyblue',
    edgecolor='black'
)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
for i in [1, 2, 3]:
    plt.axvline(mean + i * std, linestyle=':', label=f'+{i}σ')
    plt.axvline(mean - i * std, linestyle=':', label=f'-{i}σ')

```

```python
left1 = mean - 1 * std
left2 = mean - 2 * std
left3 = mean - 3 * std
right1 = mean + 1 * std
right2 = mean + 2 * std
right3 = mean + 3 * std
```

```python
df[(df.height > left1) & (df.height < right1)].describe()
```

```python
df[(df.height > left2) & (df.height < right2)].describe()
```

```python
df[(df.height > left3) & (df.height < right3)].describe()
```
