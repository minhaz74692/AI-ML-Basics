```python
import pandas as pd
```

```python
df = pd.read_csv("heights_with_outliers.csv")
df.describe()
```

```python
mean = df['height'].mean()
std = df['height'].std()
round(mean, 2), round(std,2 )
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python


sns.histplot(
    df['height'],
    bins=30,
    kde=True,
    stat='density',
    color='skyblue',
    edgecolor='black'
)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
for i in [1, 2, 3]:
    plt.axvline(mean + i * std, linestyle=':', label=f'+{i}σ')
    plt.axvline(mean - i * std, linestyle=':', label=f'-{i}σ')
```

```python
lower = mean - 3 * std
upper = mean + 3* std
lower, upper
```

```python
df_clean = df[(df.height > lower) & (df.height < upper)]
mean_clean = df_clean['height'].mean()
std_clean = df_clean['height'].std()
round(mean_clean, 2), round(std_clean,2 )
```

```python

sns.histplot(
    df_clean['height'],
    bins=30,
    kde=True,
    stat='density',
    color='skyblue',
    edgecolor='black'
)
plt.axvline(mean, color='red', linestyle='--', label='Mean')
for i in [1, 2, 3]:
    plt.axvline(mean_clean + i * std_clean, linestyle=':', label=f'+{i}σ')
    plt.axvline(mean_clean - i * std_clean, linestyle=':', label=f'-{i}σ')
```
