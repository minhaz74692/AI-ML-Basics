**Stack**

```python
c= np.array([
    [101, "Minhaz", "SWE"],
    [102, "Monir", "CE"],
    [103, "Kamrul", "TA"]
])

d= np.array([
    [101, 28, "Sylhet"],
    [102, 31, "Dhaka"],
    [103, 34, "CTG"]
])
```

```python
np.hstack((c,d))
```

```python
e= np.array([
    [104, "Rahul", "SWE"],
    [105, "Harun", "CE"],
])
```

```python
f = np.vstack((c,e))
```

```python
f
```

```python
np.hsplit(f, [2])
```

```python
np.vsplit(f, [2])
```
