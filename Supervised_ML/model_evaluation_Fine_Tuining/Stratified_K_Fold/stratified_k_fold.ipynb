```python
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_classes=2,
    n_informative=8,
    n_redundant=2,
    n_repeated=0,
    weights=[0.9, 0.1], #Which actually indicating class imbalance
    random_state=42
)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from collections import Counter
print(Counter(y))
```

## Implement K-Fold

```python
from sklearn.model_selection import KFold

kf = KFold(n_splits=5, random_state=42, shuffle=True)
for train_index, test_index in kf.split(X,y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print("Train --->", Counter(y_train))
    print("Test --->", Counter(y_test), "\n")
```

```python
from sklearn.model_selection import StratifiedKFold
skf = StratifiedKFold(n_splits=5, random_state=42, shuffle=True)
for train_index, test_index in skf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]
    print("Train --->", Counter(y_train))
    print("Test --->", Counter(y_test), "\n")
```

# Find the scores by XGBoost

```python
from sklearn.model_selection import cross_val_score
from xgboost import XGBClassifier
kf_score = cross_val_score(XGBClassifier(), X, y, cv=kf)
skf_scores = cross_val_score(XGBClassifier(), X, y, cv=skf)

print("K-Fold Scores: --->", kf_score)
print("Stratified-K-Fold Scores: --->", skf_scores)
```
