```python
import pandas as pd
```

```python
df = pd.read_excel("Raisin_Dataset.xlsx")
```

```python
df.head()
```

## Train Test Split

```python
X = df[["Area", "MajorAxisLength", "MinorAxisLength", "Eccentricity", "ConvexArea", "Extent", "Perimeter"]]
y = df["Class"]
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
X_train.shape
```

## Scale the Data

```python
from sklearn.preprocessing import MinMaxScaler, StandardScaler
scaler = StandardScaler()
scaler.fit(X)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

## Model Training using SVM: RBF Kernel: No Scaling

```python
from sklearn.svm import SVC

model =  SVC(kernel="rbf")
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# model.score(y_test, y_pred)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

print("Number of iteration:", model.n_iter_)
print(report)
```

## Model Training using SVM: RBF Kernel: With Scaling

```python
from sklearn.svm import SVC

model =  SVC(kernel="rbf")
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

# model.score(y_test, y_pred)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

print("Number of iteration:", model.n_iter_)
print(report)
```

## Model Training using SVM: Linear Kernel: NO Scaling

```python
from sklearn.svm import SVC

model =  SVC(kernel="linear")
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

# model.score(y_test, y_pred)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

print("Number of iteration:", model.n_iter_)
print(report)
```

## Model Training using SVM: Linear Kernel: With Scaling

```python
from sklearn.svm import SVC

model =  SVC(kernel="linear")
model.fit(X_train_scaled, y_train)

y_pred = model.predict(X_test_scaled)

# model.score(y_test, y_pred)

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)

from sklearn.metrics import classification_report
report = classification_report(y_test, y_pred)

print("Number of iteration:", model.n_iter_)
print(report)
```
