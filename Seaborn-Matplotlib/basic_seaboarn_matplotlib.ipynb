```python
import pandas as pd
```

```python
df = pd.read_excel("linechart.xlsx")
```

```python
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
plt.figure(figsize=(12,5))
plt.plot(df.Quarter, df.Fridge, color = "blue", label="fridge")
plt.plot(df.Quarter, df.Dishwasher, color = "green", label = "dishwasher")
plt.plot(df.Quarter, df["Washing Machine"], color = "yellow", label = "washing machine")
plt.ylabel("revenue(in $)")
plt.xlabel("quarter")
plt.legend()
plt.show()
```

```python
total_sales = df[["Fridge", "Dishwasher", "Washing Machine"]].sum()
plt.pie(total_sales, labels=total_sales.index, autopct="%1.2f%%", explode=(0.1,.2,0), shadow=True, startangle=180)

plt.show()
```
