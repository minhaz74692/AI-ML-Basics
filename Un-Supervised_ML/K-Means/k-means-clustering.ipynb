```python
import pandas as pd
from matplotlib.pyplot import grid
```

```python
df = pd.read_csv("income.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(df, x="Age", y="Income($)")
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=0)
kmeans.fit(df[["Age", "Income($)"]])
```

```python
kmeans.labels_
```

```python
df["Cluster"] = kmeans.labels_
```

```python
df.head()
```

```python
sns.scatterplot(df, x="Age", y="Income($)", hue="Cluster", palette="viridis", s=100)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaled_df = pd.DataFrame(scaler.fit_transform(df[["Age", "Income($)"]]), columns=["Age", "Income($)"])
scaled_df.head()
```

```python
scaled_kmeans = KMeans(n_clusters=3, random_state=0)
scaled_kmeans.fit(scaled_df[["Age", "Income($)"]])
scaled_df["Cluster"] = scaled_kmeans.labels_
scaled_df.head()
centroid = scaled_kmeans.cluster_centers_
centroid
```

```python
sns.scatterplot(scaled_df, x="Age", y="Income($)", hue="Cluster", s=100, palette="viridis")
sns.scatterplot( x = centroid[:,0], y= centroid[:,1], marker="*",  s=100, color="red")
```

## Identify K value with Elbow technique

```python
sse = []
k_rng = range(1,10)

for k in k_rng:
    km = KMeans(n_clusters=k, random_state=0)
    km.fit(scaled_df[["Age", "Income($)"]])
    sse.append(km.inertia_) #Where inertia is indicating the SSE

plt.plot(k_rng, sse)
```
