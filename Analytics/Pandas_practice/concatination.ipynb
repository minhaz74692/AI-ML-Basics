```python
import pandas as pd
```

```python
india_df = pd.DataFrame({
    "city": ["mumbai", "delhi", "chennai"],
    "temperature": [23,15, 25],
    "humidity": [65,24,62]
})
```

```python
us_df = pd.DataFrame({
    "city": ["new york", "chicago", "orlando"],
    "temperature": [12,32, 14],
    "humidity": [65,24,45]
})
```

```python
pd.concat([india_df, us_df], keys=["india", "us"])
```
