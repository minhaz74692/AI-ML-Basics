```python
#Custom Group function
def grouper(dframe,idx, col):
    if 80 <= dframe[col].loc[idx] <= 90:
        return '80-90'
    elif 50 <= dframe[col].loc[idx] <= 60:
        return "50-60"
    return 'others'
```

```python

gnew = df.groupby(lambda idx: grouper(df, idx, "temperature"))
```

```python
for key, data in gnew:
    print("key:", key)
    print("data:", data)
    print("\n")
```
